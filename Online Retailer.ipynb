## Bar Chart - Plot sales across each product category

```python
# importing the libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating two numpy arrays 
product_categories = np.array(['Furniture', 'Technology', 'Office Supplies'])
sales = np.array ([4110451.90, 4744557.50, 3787492.52])
print(type(product_categories))
print(type(sales))
```

```python
# to plot the bar graph
# plt.bar(x-axis, y-axis)
plt.bar(product_categories, sales)
plt.show()
```

```python
plt.bar(product_categories, sales, width=0.3, color='red', align='center', edgecolor='blue')
plt.title("Product Categories Sale\n", fontdict={'fontsize':15, 'fontweight':5, 'color':'blue'})

# Labeling Axes
plt.xlabel("Product Category", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})
plt.ylabel("Sales", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})

# Modifying the ticks to show information in (million dollars)
ticks = np.arange(0, 6000000, 1000000)
labels = ["{}M".format(i//1000000) for i in ticks]
plt.yticks(ticks, labels)
plt.show()
```
